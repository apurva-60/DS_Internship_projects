### Split the data into training and testing sets

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop(columns=['Class'],axis = 1)
Y = df['Class']
```

```python
X
```

```python
Y
```

```python
X.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.30)
```

```python
from sklearn.tree import DecisionTreeClassifier

dtree = DecisionTreeClassifier(criterion='entropy', random_state=0)

```

```python
dtree.fit(X_train,y_train)
```

### Prediction 

```python
predictions = dtree.predict(X_test)
```

```python
predictions
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,predictions))
```

```python
print(confusion_matrix(y_test,predictions))
```

```python
from sklearn import tree
plt.figure(figsize=(20,25))
tree.plot_tree(dtree,feature_names=X.columns,class_names=['Class-1', 'Class-0'],rounded=True, # Rounded node edges
          filled=True, # Adds color according to class
          proportion=True
        )
plt.show()
```

### Evaluate the model on the training and testing datasets

```python
from sklearn.metrics import accuracy_score
```

```python
prediction_on_training_data = dtree.predict(X_train)
```

```python
accuracy_of_training_data = accuracy_score(y_train,prediction_on_training_data)
```

```python
prediction_on_test_data = dtree.predict(X_test)
accuracy_on_test_data = accuracy_score(y_test,prediction_of_test_data)
```

```python
accuracy_of_training_data
```

```python
accuracy_of_test_data
```

### Visualization

```python
# Create a bar chart to visualize model accuracy
labels = ['Training Data', 'Test Data']
accuracy_values = [accuracy_of_training_data, accuracy_of_test_data]

plt.bar(labels, accuracy_values)
plt.title('Model Accuracy on Training and Test Data')
plt.ylabel('Accuracy',c="r")
plt.ylim(0, 1.0)
plt.show()
```
